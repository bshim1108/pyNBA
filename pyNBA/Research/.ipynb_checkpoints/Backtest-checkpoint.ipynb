# Backtest
In this section, I will backtest an investment strategy that enters Draftkings DFS contests based on my player statistic projections. I *temporarily* limit the backtest to Draftkings contests due to the current lack of data on the historical results of DFS contests on other sites. The methods that determine which contests I will enter and the lineups that enter each contest will be explored in this notebook. 

## Setup

Import the necessary libraries.

```python
# for data
from pyNBA.Data.constants import ROTO_NAME_TO_NBA_NAME, DB_TEAM_TO_NBA_TEAM, BAD_CONTEST_SUBSTRINGS
from pyNBA.Data.data import QueryData
from pyNBA.Models.helpers import CleanData
import pandas as pd
import numpy as np
from numpy.random import normal

# for models
from pyNBA.DFS.rules import FPCalculator
from pyNBA.DFS.constants import Site
from pyNBA.Models.fantasy_points import FPModel
from pyNBA.Models.variance import VarianceModel

# for backtesting
import time
from pydfs_lineup_optimizer import get_optimizer, Sport
from pydfs_lineup_optimizer import Player

# for plotting
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
from datetime import datetime
import seaborn as sns; sns.set()

# misc
import warnings
warnings.filterwarnings('ignore')
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

## Data
Collect all data necessary to perform the backtest.

```python
%%time

query_data = QueryData()
clean_data = CleanData()

# historical player boxscore data
boxscores = query_data.query_boxscore_data()

# we are only interested in the boxscores of healthy and non-resting players
boxscores = clean_data.drop_rows_player_injured(boxscores)
boxscores = clean_data.drop_rows_player_rest(boxscores)

# historical quarterly boxscore data
quarterly_boxscore_data = query_data.query_quarterly_boxscore_data()

# historical vegas odds data
odds_data = query_data.query_odds_data()

# historical DFS salary data
salaries = query_data.query_salary_data()
salaries = salaries.rename(columns={"POSITION": "DFS_POSITION"})

# historical DFS contest data. Strategy currently only optimized for Classic contests
contests = query_data.query_contest_data()
contests = contests.loc[
    (contests['CASHLINE'] > 0) & (contests['SLATETYPE'] == 'Classic') & (contests['CASHLINE'] > 200) &
    (~contests['CONTESTNAME'].str.lower().str.contains('|'.join(BAD_CONTEST_SUBSTRINGS)))
].dropna(subset=['CASHLINE'])

# contest payout structure
contest_info = query_data.query_contest_info_data()
```

Merge boxscore and salary data.

```python
def name_converter(name):
    name_list = name.split(',')
    name = "{} {}".format(name_list[-1].lstrip(), ' '.join(name_list[:-1]))
    if name in ROTO_NAME_TO_NBA_NAME:
        return ROTO_NAME_TO_NBA_NAME[name]
    return name

salaries['NAME'] = salaries['PLAYER'].apply(name_converter)
boxscores = boxscores.merge(
    salaries, how='left', on=['DATE', 'NAME']
)
boxscores['REB'] = boxscores['DREB'] + boxscores['OREB']
```

## Strategy
Determine viable contests and how to enter each contest.

Initilize data.

```python
season_list = list(boxscores['SEASON'].unique())

# to calculate fantasy points
DKFPCalculator = FPCalculator(Site.DRAFTKINGS)

# isolate Draftkings salaries
dk_boxscores = boxscores.loc[boxscores['SITE'] == Site.DRAFTKINGS]

# optimize Draftkings NBA lineups
optimizer = get_optimizer(Site.DRAFTKINGS, Sport.BASKETBALL)
```

Method to get the prize in a contest given a score.

```python
def get_contest_prize(contest_info, score):
    prize = contest_info.loc[contest_info['MINPOINTS'] <= score, 'PRIZE'].max()
    return prize
```

Enter contests.

Before entering contests, I must generate fantasy point projects. I will generate these projections one season at a time, using all previous seasons to train the model. Then, I will iterate through each slate in the season and enter each contest in each slate with the optimal lineup determined by my predictions. I choose to enter every contest because I will use this initial contest entry backtest to determine which contests are the most profitable (for my projections). In the next backtest, I will isolate the viable contests and enter those contests more strategically.

```python
print(time.ctime())

results = pd.DataFrame(columns = [
    'DATE', 'GAME_COUNT', 'CONTEST_TYPE', 'MAX_ENTRIES', 'TOTAL_ENTRIES', 'ENTRY_FEE', 'CASHLINE', 'PRED_SCORE', 'SCORE', 'PRIZE', 'TOP_PRIZE'
])

for i in range(1, len(season_list)):
    print(season_list[0:i], season_list[i])
    train_seasons = dk_boxscores.loc[dk_boxscores['SEASON'].isin(season_list[0:i])]
    test_season = dk_boxscores.loc[dk_boxscores['SEASON'] == season_list[i]]
    
    # check if I have contest data for this season
    season_contests = contests.loc[
        (contests['DATE'] >= test_season['DATE'].min()) & (contests['DATE'] <= test_season['DATE'].max())
    ]
    if season_contests.empty:
        continue

    # actual fantasy points
    test_season['DKFP'] = test_season.apply(
        lambda x: DKFPCalculator.calculate_fantasy_points(
            x['SEASON'], x['PTS'], x['REB'], x['AST'], x['TOV'], x['BLK'], x['STL'], x['FG3M']
        ), 
        axis=1
    )

    # project fantasy points
    fp_model = FPModel(train_seasons, test_season, Site.DRAFTKINGS)
    fp_model.train_model(quarterly_boxscore_data, odds_data)
    predictions, prediction_column = fp_model.predict()
    predictions = predictions.dropna(subset=[prediction_column])
    print(time.ctime())

    # iterate through each slate in the season
    for (date, slate), daily_contests in season_contests.groupby(['DATE', 'SLATEID']):
        # players in slate
        slate_teams = daily_contests.iloc[0]['TEAMS'].split('_')
        slate_teams = [i if i not in DB_TEAM_TO_NBA_TEAM else DB_TEAM_TO_NBA_TEAM[i] for i in slate_teams]
        board = predictions.loc[
            (predictions['DATE'] == date) & (predictions['TEAM'].isin(slate_teams))
        ]
        
        # check if we have enough predictions to generate a lineup
        pred_teams = list(board['TEAM'].unique())
        game_count = daily_contests.iloc[0]['GAMECOUNT']
        if len(pred_teams)/2 != game_count:
            continue

        # choose contests to enter
        daily_contests_to_enter = daily_contests

        for _, contest_to_enter in daily_contests_to_enter.iterrows():
            result = {}    
            result['DATE'] = date
            result['GAME_COUNT'] = game_count
            result['MAX_ENTRIES'] = contest_to_enter['MAXENTRIES']
            result['TOTAL_ENTRIES'] = contest_to_enter['TOTALENTRIES']
            result['ENTRY_FEE'] = contest_to_enter['ENTRYFEE']
            result['CASHLINE'] = contest_to_enter['CASHLINE']
            
            # contest_type
            max_roi = contest_to_enter['TOPPRIZE']/contest_to_enter['ENTRYFEE']
            if max_roi == 2:
                contest_type = 'DOUBLE_UP'
            elif max_roi > 2:
                contest_type = 'TOURNAMENT'
            else:
                contest_type = 'UNKNOWN'
            result['CONTEST_TYPE'] = contest_type

            # contest prize thresholds
            contest_id = contest_to_enter['CONTESTID']
            entered_contest_info = contest_info.loc[contest_info['CONTESTID'] == contest_id]
            if entered_contest_info.empty:
                continue

            # optiminze n lineups
            players = []
            for player_id, player in board.groupby(['PLAYERID']):
                player = player.iloc[0]
                name = player['NAME'].split()
                first_name = name[0]
                last_name = name[1] if len(name) > 1 else ''
                player = Player(
                    player_id, first_name, last_name, player['DFS_POSITION'].split('_'),
                    player['TEAM'], player['SALARY'], player[prediction_column]
                )
                players.append(player)
            optimizer.load_players(players)
            lineups = optimizer.optimize(n=1)

            # enter lineups into contest
            entered_contest_info['MINPOINTS'] = entered_contest_info['MINPOINTS'].fillna(0)
            
            for l in lineups:
                player_ids = [i.id for i in l.lineup]
                score = board.loc[board['PLAYERID'].isin(player_ids), 'DKFP'].sum()
                pred_score = board.loc[board['PLAYERID'].isin(player_ids), prediction_column].sum()
                result['PRED_SCORE'] = round(pred_score, 2)
                result['SCORE'] = score

                prize = get_contest_prize(entered_contest_info, score)
                result['PRIZE'] = prize
                result['TOP_PRIZE'] = contest_to_enter['TOPPRIZE']
                
                results = results.append(result, ignore_index=True)
```

```python
fig = plt.figure(1, figsize=(40, 30))
gs = fig.add_gridspec(2, 2)

# overall PNL
temp = results.copy()
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['RUNNING_PNL'] = pnl['PNL'].cumsum()

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('Overall PNL', fontsize=28)
pnl['COLOR'] = pnl['PNL'].apply(lambda x: 'green' if x > 0 else 'red')
sns.barplot(x=pnl.index, y='PNL', data=pnl, palette=pnl['COLOR'], ax=ax1)

xcoords = {
    pnl.index.min(): ('2017-18 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2018-02-08'][0]: ('2017-18 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2018-02-22'][0]: ('2017-18 SEASON ALL-STAR RETURN', 'y'),
    pnl.index[pnl['DATE'] == '2018-10-19'][0]: ('2018-19 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2019-02-07'][0]: ('2018-19 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2019-02-21'][0]: ('2018-19 SEASON ALL-STAR RETURN', 'y'),
    pnl.index[pnl['DATE'] == '2019-10-25'][0]: ('2019-20 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2020-02-06'][0]: ('2019-20 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2020-02-20'][0]: ('2019-20 SEASON ALL-STAR RETURN', 'y'),
    
}
for xcoord in xcoords:
    ax1.axvline(x=xcoord, label=xcoords[xcoord][0], color=xcoords[xcoord][1], linestyle='dashed')
ax1.legend(bbox_to_anchor=(1, 1.35), loc='upper right', prop={'size': 16})

ax2 = ax1.twinx()
sns.lineplot(x=pnl.index, y='RUNNING_PNL', data=pnl, sort=False, color='tab:blue', ax=ax2)

ax1.set_xlabel('Date', fontsize=22)
ax1.set_ylabel('Daily PNL', fontsize=22)
ax2.set_ylabel('Cumulative PNL', fontsize=22)
ax1.set_xticks([])

ax1.tick_params(axis='both', labelsize=16)
ax2.tick_params(axis='both', labelsize=16)
    
# cash PNL
temp = results.loc[results['CONTEST_TYPE'] == 'DOUBLE_UP']
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['RUNNING_PNL'] = pnl['PNL'].cumsum()

ax3 = fig.add_subplot(gs[1, 0])
ax3.set_title('Cash Game PNL', fontsize=28)
pnl['COLOR'] = pnl['PNL'].apply(lambda x: 'green' if x > 0 else 'red')
sns.barplot(x=pnl.index, y='PNL', data=pnl, palette=pnl['COLOR'], ax=ax3)

ax4 = ax3.twinx()
sns.lineplot(x=pnl.index, y='RUNNING_PNL', data=pnl, sort=False, color='tab:blue', ax=ax4)

ax3.set_xlabel('Date', fontsize=22)
ax3.set_ylabel('Daily PNL', fontsize=22)
ax4.set_ylabel('Cumulative PNL', fontsize=22)
ax3.set_xticks([])

ax3.tick_params(axis='both', labelsize=16)
ax4.tick_params(axis='both', labelsize=16)

# tournament PNL
temp = results.loc[results['CONTEST_TYPE'] == 'TOURNAMENT']
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['RUNNING_PNL'] = pnl['PNL'].cumsum()

ax5 = fig.add_subplot(gs[1, 1])
ax5.set_title('Tournament PNL', fontsize=28)
pnl['COLOR'] = pnl['PNL'].apply(lambda x: 'green' if x > 0 else 'red')
sns.barplot(x=pnl.index, y='PNL', data=pnl, palette=pnl['COLOR'], ax=ax5)

ax6 = ax5.twinx()
sns.lineplot(x=pnl.index, y='RUNNING_PNL', data=pnl, sort=False, color='tab:blue', ax=ax6)

ax5.set_xlabel('Date', fontsize=22)
ax5.set_ylabel('Daily PNL', fontsize=22)
ax6.set_ylabel('Cumulative PNL', fontsize=22)
ax5.set_xticks([])

ax5.tick_params(axis='both', labelsize=16)
ax6.tick_params(axis='both', labelsize=16)

plt.show()
```

It is important not to overfit by cherry picking the contests in which the projections made money. To try to avoid this, I bucketed the entry fees and slate game counts. I will also only isolate a subset of contests if it makes sense why they were more profitable. The results above show that virtually no subset of cash games are profitable. On the other hand, tournaments whose entry fees are at most $1000 and slate game counts are at most 6 look profitable. This makes sense: higher entry contests attract more intelligent players, and high game slates introduce more variance.

Below I will graph the cumulative PNL of a strategy that only enters contests in the subset described above. I will also include cash games whose entry fees are at most $25 and slate game counts are at most 3 to verify the lack of edge.

```python
fig = plt.figure(1, figsize=(40, 15))
gs = fig.add_gridspec(1, 2)
    
# cash PNL
temp = results.loc[(results['CONTEST_TYPE'] == 'DOUBLE_UP') & (results['ENTRY_FEE'] <= 25) & (results['GAME_COUNT'] <= 3)]
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['RUNNING_PNL'] = pnl['PNL'].cumsum()

ax1 = fig.add_subplot(gs[0, 0])
ax1.set_title('Cash Game PNL', fontsize=28)
pnl['COLOR'] = pnl['PNL'].apply(lambda x: 'green' if x > 0 else 'red')
sns.barplot(x=pnl.index, y='PNL', data=pnl, palette=pnl['COLOR'], ax=ax1)

ax2 = ax1.twinx()
sns.lineplot(x=pnl.index, y='RUNNING_PNL', data=pnl, sort=False, color='tab:blue', ax=ax2)

ax1.set_xlabel('Date', fontsize=22)
ax1.set_ylabel('Daily PNL', fontsize=22)
ax2.set_ylabel('Cumulative PNL', fontsize=22)
ax1.set_xticks([])

ax1.tick_params(axis='both', labelsize=16)
ax2.tick_params(axis='both', labelsize=16)

# tournament PNL
temp = results.loc[(results['CONTEST_TYPE'] == 'TOURNAMENT') & (results['ENTRY_FEE'] <= 500) & (results['GAME_COUNT'] <= 6)]
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['RUNNING_PNL'] = pnl['PNL'].cumsum()

ax3 = fig.add_subplot(gs[0, 1])
ax3.set_title('Tournament PNL', fontsize=28)
pnl['COLOR'] = pnl['PNL'].apply(lambda x: 'green' if x > 0 else 'red')
sns.barplot(x=pnl.index, y='PNL', data=pnl, palette=pnl['COLOR'], ax=ax3)

xcoords = {
    pnl.index.min(): ('2017-18 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2018-02-08'][0]: ('2017-18 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2018-02-22'][0]: ('2017-18 SEASON ALL-STAR RETURN', 'y'),
    pnl.index[pnl['DATE'] == '2018-10-19'][0]: ('2018-19 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2019-02-07'][0]: ('2018-19 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2019-02-21'][0]: ('2018-19 SEASON ALL-STAR RETURN', 'y'),
    pnl.index[pnl['DATE'] == '2019-10-25'][0]: ('2019-20 SEASON START', 'k'),
    pnl.index[pnl['DATE'] == '2020-02-06'][0]: ('2019-20 SEASON TRADE DEADLINE', 'm'),
    pnl.index[pnl['DATE'] == '2020-02-20'][0]: ('2019-20 SEASON ALL-STAR RETURN', 'y'),
    
}
for xcoord in xcoords:
    ax3.axvline(x=xcoord, color=xcoords[xcoord][1], linestyle='dashed')

ax4 = ax3.twinx()
sns.lineplot(x=pnl.index, y='RUNNING_PNL', data=pnl, sort=False, color='tab:blue', ax=ax4)

ax3.set_xlabel('Date', fontsize=22)
ax3.set_ylabel('Daily PNL', fontsize=22)
ax4.set_ylabel('Cumulative PNL', fontsize=22)
ax3.set_xticks([])

ax3.tick_params(axis='both', labelsize=16)
ax4.tick_params(axis='both', labelsize=16)

plt.show()
```

I will only focus on tournament contests going forward. The tournament strategy seems like it may actually be profitable. However, there is a lot of variance in PNL, as the cumulative PNL depends entirely on a few days in which it hit big. Most days the strategy loses money, and there are some significant drawdowns.

Below shows the distribution of daily returns of this strategy.

```python
temp = results.loc[(results['CONTEST_TYPE'] == 'TOURNAMENT') & (results['ENTRY_FEE'] <= 500) & (results['GAME_COUNT'] <= 6)]
temp['PNL'] = temp['PRIZE'] - temp['ENTRY_FEE']
pnl = temp.groupby(['DATE']).apply(
    lambda x: pd.Series({
        'ENTRY_FEE': x['ENTRY_FEE'].sum(),
        'PRIZE': x['PRIZE'].sum(),
        'PNL': x['PNL'].sum()
    })
).reset_index()
pnl['ROI'] = pnl['PNL']/pnl['ENTRY_FEE']
ax = sns.distplot(pnl['ROI'])
ax.set_title('PNL Distribution', fontsize=18)
plt.rcParams["figure.figsize"] = (20, 10)
plt.show()
```

Currently, the most common outcome by far is a -100% ROI. The goal is to make this distribution more consistently positive.

I will reduce the variance in returns by diversifying the lineups I enter in each slate. In each tournament contest that meets the criterion described above, I will generate as many different lineups as the contest allows me to enter. To ensure diversity in the set of lineups I enter, I will change the nature of my projections from a constant number to a distribution of possibilities. The mean of the distribution will be the projected fantasy points. The standard deviation of the distribution for player $P$ in season $S$ will be the mean residual of my projection for $P$ in $S$. Moreover, I will cap the exposure to a player in a contest at 80%. This number is currently arbitrary. In the future, I will project ownership percentages for each player, and cap my exposure to each player at a function of the projected ownership of that player.

```python
print(time.ctime())

randomness_results = pd.DataFrame(columns = [
    'DATE', 'GAME_COUNT', 'CONTEST_TYPE', 'CONTEST_NAME', 'MAX_ENTRIES', 'TOTAL_ENTRIES', 'ENTRY_FEE', 'CASHLINE', 'SCORE', 'PRIZE', 'TOP_PRIZE'
])
pnl = 0
for i in range(1, len(season_list)):
    print(season_list[0:i], season_list[i])
    train_seasons = dk_boxscores.loc[dk_boxscores['SEASON'].isin(season_list[0:i])]
    test_season = dk_boxscores.loc[dk_boxscores['SEASON'] == season_list[i]]
    
    # check if I have contest data for this season
    season_contests = contests.loc[
        (contests['DATE'] >= test_season['DATE'].min()) & (contests['DATE'] <= test_season['DATE'].max())
    ]
    if season_contests.empty:
        continue

    # actual fantasy points
    test_season['DKFP'] = test_season.apply(
        lambda x: DKFPCalculator.calculate_fantasy_points(
            x['SEASON'], x['PTS'], x['REB'], x['AST'], x['TOV'], x['BLK'], x['STL'], x['FG3M']
        ), 
        axis=1
    )

    # project fantasy points
    fp_model = FPModel(train_seasons, test_season, Site.DRAFTKINGS)
    fp_model.train_model(quarterly_boxscore_data, odds_data)
    predictions, prediction_column = fp_model.predict()
    predictions = predictions.dropna(subset=[prediction_column])
    
    # variance
    variance_model = VarianceModel(predictions)
    predictions, variance_column = variance_model.predict('DKFP', prediction_column)
    pct_variance_column = 'RESIDUAL_PCT'
    predictions[pct_variance_column] = predictions[variance_column]/predictions[prediction_column]
    predictions[pct_variance_column] = predictions[pct_variance_column].replace([np.inf, -np.inf], np.nan)
    predictions[pct_variance_column] = predictions[pct_variance_column].fillna(0.25)
    predictions[pct_variance_column] = predictions[pct_variance_column].clip(lower=0.05)
    
    print(time.ctime())

    # iterate through each slate in the season
    for (date, slate), daily_contests in season_contests.groupby(['DATE', 'SLATEID']):
        print(date)
        # players in slate
        slate_teams = daily_contests.iloc[0]['TEAMS'].split('_')
        slate_teams = [i if i not in DB_TEAM_TO_NBA_TEAM else DB_TEAM_TO_NBA_TEAM[i] for i in slate_teams]
        board = predictions.loc[
            (predictions['DATE'] == date) & (predictions['TEAM'].isin(slate_teams))
        ]

        # check if we have enough predictions to generate a lineup
        pred_teams = list(board['TEAM'].unique())
        game_count = daily_contests.iloc[0]['GAMECOUNT']
        if len(pred_teams)/2 != game_count:
            continue

        # choose contests to enter
        daily_contests['MAXROI'] = daily_contests['TOPPRIZE']/daily_contests['ENTRYFEE']
        daily_contests_to_enter = daily_contests.loc[(daily_contests['MAXROI'] > 2) & (daily_contests['ENTRYFEE'] <= 500)]

        for _, contest_to_enter in daily_contests_to_enter.iterrows():
            contest_name = contest_to_enter['CONTESTNAME']
            print(contest_name)

            result = {}    
            result['DATE'] = date
            result['GAME_COUNT'] = game_count
            result['MAX_ENTRIES'] = contest_to_enter['MAXENTRIES']
            result['TOTAL_ENTRIES'] = contest_to_enter['TOTALENTRIES']
            result['ENTRY_FEE'] = contest_to_enter['ENTRYFEE']
            result['CASHLINE'] = contest_to_enter['CASHLINE']
            result['CONTEST_TYPE'] = 'TOURNAMENT'
            result['CONTEST_NAME'] = contest_name

            # contest prize thresholds
            contest_id = contest_to_enter['CONTESTID']
            entered_contest_info = contest_info.loc[contest_info['CONTESTID'] == contest_id]
            if entered_contest_info.empty:
                continue
            entered_contest_info['MINPOINTS'] = entered_contest_info['MINPOINTS'].fillna(0)

            # randomness and exposure
            try:
                players = []
                for player_id, player in board.groupby(['PLAYERID']):
                    player = player.iloc[0]
                    name = player['NAME'].split()
                    first_name = name[0]
                    last_name = name[1] if len(name) > 1 else ''
                    mu = player[prediction_column]
                    sigma = player[pct_variance_column]
                    salary =  player['SALARY']
                    if (mu * (1+ normal(0, sigma)))/salary > 0.006 or (mu * (1+sigma))/salary > 0.006:
                        player = Player(
                            player_id, first_name, last_name, player['DFS_POSITION'].split('_'),
                            player['TEAM'], salary, mu, standard_deviation=sigma/2
                        )
                        players.append(player)
                optimizer.load_players(players)
                optimizer.set_min_salary_cap(49500)
                lineups = optimizer.optimize(n=int(contest_to_enter['MAXENTRIES']), randomness=True)

                for l in lineups:
                    player_ids = [i.id for i in l.lineup]
                    score = board.loc[board['PLAYERID'].isin(player_ids), 'DKFP'].sum()
                    result['SCORE'] = score

                    prize = get_contest_prize(entered_contest_info, score)
                    result['PRIZE'] = prize
                    result['TOP_PRIZE'] = contest_to_enter['TOPPRIZE']
                    randomness_results = randomness_results.append(result, ignore_index=True)

                    p = (prize - contest_to_enter['ENTRYFEE'])
                    if not np.isnan(p):
                        pnl += p
                        print('TOT:', round(pnl), round(prize), round(contest_to_enter['ENTRYFEE']), round(contest_to_enter['TOPPRIZE']))
            except:
                print('FAILURE')
                continue
```
