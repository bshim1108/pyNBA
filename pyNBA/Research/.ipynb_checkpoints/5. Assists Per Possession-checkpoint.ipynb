# Data Analysis
Summarize the characteristics of the dataset.

```python
query_data = QueryData()
clean_data = CleanData()
helpers = Helpers()

# pull boxscore data
boxscores = query_data.query_boxscore_data()
boxscores = clean_data.select_regular_season_games(boxscores)
boxscores = clean_data.drop_rows_player_injured(boxscores)
boxscores = clean_data.drop_rows_player_rest(boxscores)
```

```python
boxscores['POSSESSIONS'] = boxscores['PACE'] * (boxscores['SECONDSPLAYED']/(48*60))
boxscores['ASSISTS/POSSESSION'] = boxscores['AST']/boxscores['POSSESSIONS']
```

# Feature Engineering
Create new features from raw data using domain knowlege.

```python
feature_creation = FeatureCreation()

boxscores = feature_creation.expanding_weighted_mean(
    df=boxscores, group_col_names=['SEASON', 'TEAM', 'PLAYERID'], col_name='ASSISTS/POSSESSION', new_col_name='AVG_ASSISTS/POSSESSION',
    weight_col_name='POSSESSIONS'
)
```

```python
boxscores = feature_creation.expanding_sum(
    df=boxscores, group_col_names=['SEASON', 'TEAM', 'PLAYERID'], col_name='POSSESSIONS', new_col_name='SUM_POSSESSIONS'
)
sig_boxscores = boxscores.loc[(boxscores['POSSESSIONS'] > 5) & (boxscores['SUM_POSSESSIONS'] > 500)]

y = sig_boxscores['ASSISTS/POSSESSION']
y_hat = sig_boxscores['AVG_ASSISTS/POSSESSION']

# Kpi
print("R2 (explained variance):")
print("Model:", round(metrics.r2_score(y, y_hat), 3))

print("\nMean Absolute Error (Σ|y-pred|/n):")
print("Model:", round(metrics.mean_absolute_error(y, y_hat), 6))

print("\nRoot Mean Squared Error (sqrt(Σ(y-pred)^2/n)):")
print("Model:", round(np.sqrt(metrics.mean_squared_error(y, y_hat)), 6))
```

```python
residuals = y - y_hat

# plot predicted vs true
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 10))
sns.scatterplot(y_hat, y, ax=ax[0])
abline_plot(intercept=0, slope=1, color="red", ax=ax[0])
ax[0].grid(True)
ax[0].set(xlabel="Predicted", ylabel="True", title="Predicted vs True")
ax[0].legend()
    
# plot predicted vs residuals
sns.scatterplot(y_hat, residuals, ax=ax[1])
ax[1].grid(True)
ax[1].set(xlabel="Predicted", ylabel="Residuals", title="Predicted vs Residuals")
ax[1].hlines(y=0, xmin=np.min(y_hat), xmax=np.max(y_hat))
ax[1].legend()
plt.show()

fig, ax = plt.subplots(figsize=(18, 5))
sns.distplot(residuals, color="red", hist=True, kde=True, kde_kws={"shade":True}, ax=ax)
ax.grid(True)
ax.set(title="Residuals distribution")
plt.show()
```
